## <center> Import Sample Data via </center>

### Import Modules

```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

### Import Data

```python
df = pd.read_csv('Data/player_info.csv')
df.head()
```

### Create a New Database

```python
# Import Postgress Connector
conn = psycopg2.connect(
    host='localhost',
    port=54320, #5432 will also work
    user='postgres',
)
```

```python
# Open connection and create database
conn.autocommit = True
cur = conn.cursor()

cur.execute("DROP DATABASE IF EXISTS nhl;")
cur.execute("CREATE DATABASE nhl;")
cur.close()
conn.close()
```

### Establish a connections to Database

```python
engine = create_engine('postgresql://postgres@localhost:54320/nhl')
table_name = 'hockey_players'

df.to_sql(table_name,con=engine, if_exists='replace', index=False, chunksize=1000)
```

### Read Data from Database

```python
 pd.read_sql_query('SELECT * FROM "hockey_players"',con=engine)
```
